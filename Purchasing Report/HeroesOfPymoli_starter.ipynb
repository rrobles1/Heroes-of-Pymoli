## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Item = len(purchase_data["Item ID"].unique())
AvgPrice = purchase_data["Price"].mean()
TotalPurchases = len(purchase_data["Purchase ID"].unique())
TotalRevenue = purchase_data["Price"].sum()
# some items have the same name but different ID. 
Purchase_Analysis = pd.DataFrame(
    {"Number of Unique Items": [Item],
    "Average Price": [AvgPrice],
    "Number of Purchases": [TotalPurchases],
    "Total Revenue": [TotalRevenue]
    })

# Use Map to format all the columns
Purchase_Analysis["Average Price"] = Purchase_Analysis["Average Price"].map("${:.2f}".format)
Purchase_Analysis["Total Revenue"] = Purchase_Analysis["Total Revenue"].map("${:,}".format)
Purchase_Analysis
```
