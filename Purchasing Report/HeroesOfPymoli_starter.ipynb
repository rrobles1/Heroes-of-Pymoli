## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
ages = pd.DataFrame(purchase_data)
ages
newages = ages.drop(['SN', 'Gender','Item ID', 'Item Name',], axis=1)
# newages
bins = [4, 8, 12, 16, 20, 24, 28, 32, 36, 40, 44, 48, 52, 56, 60 ]
group_names = ["0-3", "4-7", "8-11", "12-15", "16-19", "20-23", "24-27", "28-31", "32-35", "36-39", "40-43", "44-47", "48-51", "52+"]
newages["Age Group"] = pd.cut(newages["Age"], bins, labels=group_names)
newages
```
